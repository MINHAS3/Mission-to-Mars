```python
from splinter import Browser
from bs4 import BeautifulSoup
```

```python
executable_path = {'executable_path': 'chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
url = 'https://astrogeology.usgs.gov/search/results?q=hemisphere+enhanced&k1=target&v1=Mars'
browser.visit(url)
```

```python
html = browser.html
soup = BeautifulSoup(html, 'html.parser')
soup
```

```python
moon_dic = {}
for i in range(0, 4):
    moon_phase = soup.findAll("div", {"class" : "item"})[i].findAll("div", {"class" : "description"})[0].find("a").text
    display(moon_phase)
    
    moon_image = soup.findAll("div", {"class" : "item"})[i].findAll("img", {"class" : "thumb"})[0].get("src")
    display(moon_image)
    
    moon_dic[moon_phase] = moon_image
```
